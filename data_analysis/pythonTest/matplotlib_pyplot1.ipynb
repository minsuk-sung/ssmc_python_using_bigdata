```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
data1 = [10,14,19,20,25]
plt.plot(data1)
plt.show()
```

```python
y = [10,23,24,34,45,58] # x축은 정렬이 되어있어야함
x = [1,2,3,4,5,6]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y1 = [10,23,24,34,45,58]
y2 = [8,12,24,36,47,53]
y3 = [34,45,22,21,45,56]
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y1 = [10,23,24,34,45,58]
y2 = [8,12,24,36,47,53]
# plt.figure()가 항상 있어서 없어도 있는거처럼
plt.plot(x,y1)
plt.figure(figsize=[3,3]) # 이게 없으면 하나의 윈도우에서 그렸을텐데 있어서 따로따로 3inch 단위로
plt.plot(x,y2)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y1 = [10,23,24,34,45,58]
y2 = [8,12,24,36,47,53]
y3 = [34,45,22,21,45,56]
plt.subplot(2,2,1) # 전체를 2행2열로 분할해서 1번째쪽에
plt.plot(x,y1)
plt.subplot(2,2,2) # 전체를 2행2열로 분할해서 2번째쪽에
plt.plot(x,y2)
plt.subplot(3,1,3) # 전체를 2행2열로 분할해서 3번째쪽에
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,14,19,20,25]
plt.ylim(0,30) # y축의 범위를 정해주는거
plt.xlim(0,10)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 7, 0.1)
y = np.sin(x)
plt.plot(x,y,'r') # 'r'은 red를 의미함
plt.show()
```

```python
x = np.arange(0, 7, 0.1)
y = np.sin(x)
plt.plot(x,y,'ro') # 'ro'은 red색인 dot형태로
plt.show()
```

```python
x = np.arange(0, 7, 0.5)
y = np.sin(x)
plt.plot(x,y,'ro-') # '-'은 solid line
plt.show()
```

```python
x = np.arange(0, 7, 0.5)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1,'ro--',x,y2,'b<-') # 'r'은 red를 의미함
plt.legend(['국어','영어']) # plot한 다음에 해줘야함
plt.show()
```

```python
x = np.arange(0, 7, 0.5)
y1 = np.sin(x)
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1,'ro--') # 'r'은 red를 의미함
plt.legend(['국어','영어']) # plot한 다음에 해줘야함
#plt.text(0,0,'0점')
for x,y in zip(x,y1):
    plt.text(x,y,"%.2f"%y)
plt.show()
```
