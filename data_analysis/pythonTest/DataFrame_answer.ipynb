```python
# 텍스트 파일을 데이터프레임으로 바꿔주는 함수 = read_table
# 통으로 읽어들임
# 문제점
# 첫번째 row는 header로 들어가버림
df = pd.read_table('births.txt')
df.head()
```

```python
# 쉼표 기준으로 끊어라(sep 속성), 헤더를 사용하지 않는다.(header=None), 0번째 칼럼을 인덱스로 지정(index_col)
df = pd.read_table('births.txt', header=None, sep=',', index_col=0 )
# 칼럼명 설정
df.columns = ['남아수', '여아수']
# 인덱스명
df. index.name='년도'
df
```

```python
# 1. 남아수 총합을 구하시오.
df['남아수'].sum()
```

```python
# 2. 여아수 평균을 구하시오.
df['여아수'].mean()
```

```python
# 3. 남아수가 가장 많은 년도와 남아수를 구하시오
# df['남아수'].idxmax(), df['남아수'].max()
df[ df['남아수']==df['남아수'].max() ]['남아수']
```

```python
# 4. 2000년도 이후 데이터를 구하시오
df.loc['2000':]
```

```python
# 5. 1995~2000년도 데이터를 구하시오
df.loc['1995':'2000']
```

```python
# 6. 남아 출생이 가장 많은 top5를 구하시오(년도와 남아수)
df.sort_values( by='남아수',
               ascending=False)[:5]['남아수']
```

```python
# 7. 남아율(%) 컬럼을 추가하고 남아수가 차지하는 비율을 보여주시오 예, 50, 70 등
df2 = df
df2['남아율(%)'] = df2['남아수']/df2.sum( axis = 1)
df2
```

```python
# 8. 2000년도 이후 데이터에 대해 남아수 출생량 컬럼을 추가하고 1,800,000 이상이면 많음, 아니면 적음이라고 보여주시오.
df3 = df.loc['2000':]
df3['출생량'] = df3['남아수'].apply(lambda n: '많음' if n >= 1800000 else '적음')
df3[['남아수', '출생량']]
```

```python
# 9. 1800년대, 1900년대, 2000년대 각 남아수, 여아수 총합을 구하시오
df4 = [ df.loc['1800':'1899'].sum(), df.loc['1900':'1999'].sum(), df.loc['2000':].sum() ]
df4 = pd.DataFrame(df4, index=['1800년대', '1900년대', '2000년대'])
df4=df4[['남아수', '여아수']]
df4
```
