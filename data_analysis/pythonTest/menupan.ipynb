## 맛집 데이터 크롤링 해오기

메뉴판 닷컴 Top 1000에서

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus']=False
from bs4 import BeautifulSoup
import urllib.request as REQ
```

### URL 주소 받아오기

```python
urlbase = 'https://www.menupan.com'
urlsub = '/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'
response = REQ.urlopen(urlbase+urlsub)
soup = BeautifulSoup(response,'html.parser')
```

```python
data = []
hrefList = []
ul = soup.select_one('.rankingList .list')
for li in ul.select('li'):
    
    # 순위에 있는 데이터 크롤링
    rank = int(li.select_one('.numTop,.rankNum').string)
    listName = li.select_one('.listName').string
    listType = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    detail_info = li.select_one('.listName a')['href']
    
    
    # 상세 정보를 가져오기 위해서
    response = REQ.urlopen(urlbase+detail_info)
    soup = BeautifulSoup(response,'html.parser')
    phoneNum = soup.select_one('.tel1').string
    detailAddr = soup.select_one('.add1').string
    
    # 가져온 데이터 넣기
    data.append({
        '순위':rank,
        '상호명':listName,
        '업종':listType,
        '지역':listArea,
        '상세주소':detailAddr,
        '전화번호':phoneNum,
    })
```

```python
df = pd.DataFrame(data)
df1 = df.set_index('순위')[['상호명','업종','전화번호','지역','상세주소']]
df1
```

### 네이버 지도 API를 통해서 위도/경도 받아오기

```python
def addrToLatLng(s):
    client_id = "ppihLuN2pr20zbkL0CAJ"
    client_secret = "OJqch1hs7D"
    encText = urllib.parse.quote(s)
    url = "https://openapi.naver.com/v1/map/geocode?query=" + encText # json ���
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        aa= response_body.decode('utf-8') 
        bb = json.loads(aa)
        return bb['result']['items'][0]['point']
    else:
        return "Error Code:" + rescode
```
