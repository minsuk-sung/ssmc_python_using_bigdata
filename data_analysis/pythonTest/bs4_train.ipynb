## 웹상에서 데이터 받아오기

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import urllib.request as REQ
````

### 조선일보 RSS를 통해서 데이터 받아오기
http://rss.joins.com/ 에서 xml 데이터를 읽어오자

````python
url = 'https://rss.joins.com/joins_news_list.xml'
response = REQ.urlopen(url)
soup = BeautifulSoup(response,'html.parser')
````

tag selector를 통해서 item을 가져오자

````python
for item in soup.select('item'): # item element는 list type으로 return 된다
    print('기사제목 : ',item.title.string)
    print('기사내용 : ',item.description.string)
    print('-'*50)
````

````python
for location in soup.select('location'):
    print(location.city.string)
````

이렇게 해서 여주까지 모두 for loop 돌리기

````python
for location in soup.select('location'):
    print(location.city.string)
    print('-'*30)
    # print(type(location)) # <class 'bs4.element.Tag'>
    for data in location.select('data'): # 여기서 다시 select로 잡아줘야하나
        print(data.tmef.string, data.wf.string, data.tmn.string, data.tmx.string )
    print('-'*30)
    print('\n')
````
