## 웹상에서 데이터 받아오기

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus']=False
from bs4 import BeautifulSoup
import urllib.request as REQ
````

### 조선일보 RSS를 통해서 데이터 받아오기
http://rss.joins.com/ 에서 xml 데이터를 읽어오자

````python
url = 'https://rss.joins.com/joins_news_list.xml'
response = REQ.urlopen(url)
soup = BeautifulSoup(response,'html.parser')
````

tag selector를 통해서 item을 가져오자

````python
for item in soup.select('item'): # item element는 list type으로 return 된다
    print('기사제목 : ',item.title.string)
    print('기사내용 : ',item.description.string)
    print('-'*50)
````

이렇게 해서 여주까지 모두 for loop 돌리기

````python
wData = []
city_no = 0
for location in soup.select('location'):
    # print(type(location)) # <class 'bs4.element.Tag'>
    #city_no += 1
    for data in location.select('data'): # 여기서 다시 select로 잡아줘야하나
        wData.append({
            '도시' : location.city.string,
            #'도시번호' : city_no,
            '날짜' : data.tmef.string,
            '날씨' : data.wf.string,
            '최저기온' : int(data.tmn.string),
            '최고기온' : int(data.tmx.string)
        })
````

````python
df = pd.DataFrame(wData)
df = df[['도시','날짜','날씨','최고기온','최저기온']]
df
````

> Q5. 인천지역의 날씨데이터를 출력하시오  
단, 느낌 컬럼을 추가하고 최저기온이
  -5도 이하이면 : 매우추워요
  -4 ~ -2 면 : 추워요
  -1도 이상이면 :조금추워요
  로 같이 표시하시오

````python
df5 = df[df['도시']=='인천']
def feel(n):
    if n <= -5:
        return '매우 추워요'
    elif n>-4 and n<-2:
        return '추워요'
    else:
        return '조금 추워요'
df5['느낌'] = df5['최저기온'].apply(feel)
````

> Q9. 수원지연 -5도 이하인 데이터의 비율을 구하시오

````python
idx = len(df[df['도시']=='수원'])
idx2, v2 = df[(df['도시']=='수원')&(df['최저기온']<-5)].shape
print('비율 : ',round(idx2/idx,2)*100,'%')
````

> Q10. 최고기온과 최저기온의 차이가 가장 높은 날씨,날짜,최저기온,최고기온을 구하시오

````python
df['기온차'] = abs(df['최고기온'] - df['최저기온'])
df[df['기온차'] == df['기온차'].max()][['도시','날씨','날짜','최저기온','최고기온','기온차']]
````
