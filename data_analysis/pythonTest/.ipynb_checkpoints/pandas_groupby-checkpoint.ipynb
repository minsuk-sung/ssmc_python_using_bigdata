```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = 'hotel.xlsx'
df = pd.read_excel(data)
```

```python
df
```

각 등급별 호텔 가격을 알아보자

여기에 사용되는 집계함수는 다음과 같다
- sum : 총합
- mean : 평균
- count : 갯수
- max : 최댓값
- min : 최솟값

```python
g = df.groupby('grade')
```

```python
g.sum()
```

```python
g.mean()
```

```python
g.count()
```

```python
g.max()
```

```python
g.min()
```

```python
g.agg(['sum','count','mean']) # multicolumn
```

```python
g.agg(['sum','count'])['price']
```
