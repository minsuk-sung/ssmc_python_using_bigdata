```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

### matplotlib 연습문제

```python
import random
x = pd.Series([n for n in range(0,10)])
kor = [random.randint(60,90) for i in range(0,10)]
eng = [random.randint(50,100) for i in range(0,10)]
math = [random.randint(50,80) for i in range(0,10)]
df = pd.DataFrame([kor,eng,math],
                 index = ['국어','영어','수학'])
df = df.T
df
```

> Q1. 국어, 영어 점수에 대한 **bar chart**를 그리시오

```python
plt.xlabel('학생')
plt.ylabel('점수')
plt.title('1학년 국어 및 영어 성적')
plt.bar(x,df['국어'],width=0.2,color='r',alpha=0.5,label='국어')
plt.bar(x+0.2,df['영어'],width=0.2,color='g',alpha=0.5,label='영어')
plt.legend()
plt.show()
```

> Q2. 영어,수학 점수에 대한 **box plot**을 그리시오

```python
plt.xlabel('학생')
plt.ylabel('점수')
plt.title('1학년 영어 및 수학 성적')
plt.boxplot([df['영어'],df['수학']],
           labels = ['영어','수학'])
plt.show()
```

> Q3. 이름별 국어 점수를 **pie chart**로 그리시오

```python
plt.pie(df['국어'])
plt.title('국어점수')
plt.show()
```

> Q4. 국어,영어,수학 점수를 **line chart**로 그리시오

```python
plt.xlabel('학생')
plt.ylabel('점수')
plt.title('1학년 국어, 영어 및 수학 성적')
plt.plot(df['국어'])
plt.plot(df['영어'])
plt.plot(df['수학'])
plt.show()
```

> Q5. 수학점수에 대한 **scatter plot**을 그리고 점수를 text로 표시하시오

```python
plt.xlabel('학생')
plt.ylabel('수학 점수')
plt.title('1학년 수학점수')
plt.scatter([n for n in range(0,10)],df['수학'])
plt.show()
```

> Q6. 국어,영어,수학 점수를 **stacked bar**로 그리시오 (% 표시)

```python
plt.xlabel('학생')
plt.ylabel('전체 점수')
plt.title('1학년 점수')

x = [n for n in range(0,10)]

plt.bar(x,df['국어'],
            color='r',
            width = 0.3,
            label = '국어',
            alpha = 0.3
           )

plt.bar(x,df['영어'],
            color='g',
            width = 0.3,
            label = '영어',
            alpha = 0.3,
            bottom = df['국어']
           )

plt.bar(x,df['수학'],
            color='b',
            width = 0.3,
            label = '수학',
            alpha = 0.3,
            bottom = df['국어']+df['영어']
           )

for x_,k,e,m in zip(x,df['국어'],df['영어'],df['수학']):
    plt.text(x_,k/2,'%.2f\n(%d점)'%((k/(k+e+m)),k),va = 'center', ha = 'center')
    plt.text(x_,e/2+k,'%.2f\n(%d점)'%((e/(k+e+m)),e),va = 'center', ha = 'center')
    plt.text(x_,m/2+k+e,'%.2f\n(%d점)'%((e/(k+e+m)),m),va = 'center', ha = 'center')

plt.legend(['국어','영어','수학'])
plt.ylim(0,330)
plt.show()
```

> Q7. `births.txt`에서 데이터의 연도별 남아/여아 출생현황을 **bar chart와 line plot로 함께** 그리시오  `plt.xtics(rotation = '45')`를 적용하여 x축 데이터를 알기 좋게 할것

```python
dfb = pd.read_table('births.txt', header=None, sep=',', index_col=0)
dfb.columns = ['남아수', '여아수']
dfb.index.name='년도'
dfb2 = dfb.loc['2000':,]
x =pd.Series(dfb2.index)
plt.xticks(x,dfb2.index, rotation='45')
plt.bar(x-0.2,dfb2['남아수'],color='red',width=0.4,alpha=0.5,label='남아수')
plt.bar(x+0.2,dfb2['여아수'],color='blue',width=0.4,alpha=0.5,label='여아수')
plt.plot(x,dfb2['남아수'])
plt.plot(x,dfb2['여아수'])
plt.xlabel('연도')
plt.ylabel('출생 수')
plt.ylim(1700000,2100000)
plt.legend()
plt.show()
```

> Q8. `hotel.xlsx` 데이터를 이용하여 호텔별 가격현황을 **bar chart**로 그리시오

> Q9. 국어를 x축으로 영어를 y축으로하여 scatter를 그린 각 원에 이름을 표시하시오

```python
plt.scatter(df['국어'],df['영어'],s=100)
for i in range(len(df['국어'])):
    plt.text(df['국어'][i],df['영어'][i]-10,'국어점수 : %s\n영어점수 :%s'%(df['국어'][i],df['영어'][i]))
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```
