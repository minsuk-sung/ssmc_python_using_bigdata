```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = 'crime_in_Seoul.csv'
df = pd.read_csv(data,
                 engine='python',
                 index_col = '관서명',
                 encoding='euc-kr',
                 thousands=','
                )
```

```python
df.head()
```

### 데이터 안에 천의 자리 숫자의 콤마 없애기

read_csv() 함수의 옵션 안에 **thousands = ','**를 주어도 괜찮지만,  
**replace()** 함수와 **astype()** 함수를 통해서 콤마를 제거할 수 있다.

```python
df.replace(',','', # df['살인 발생'].replace : 정확히 2랑 일치하는걸 111로 바꿔라 
          regex = True, # 이걸 통해서 쉼표 제거가 가능함
          inplace = True
          )
```
